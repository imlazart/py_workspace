# Tuple as an argument 

## * is to catch multiple arugments

# pass arugments as tuple and sum it

```python
def add(*a):
    print(type(a))
    print(a)
    sum=0;
    for i in a:
        sum+=i;
    print(sum)

add(2,3,4,5,7)
```

```python
def add(a, b, *morevalues):
    print(type(a))
    print(a)
    sum = a + b
    for i in morevalues:
        sum += i
    print(sum)

add(2,3,4,5,7)
```

# keyword arguments

```python
def add(a, b,c=3, *args, **kwargs):
    print(type(a))
    print(a)
    print("-"*10)
    print(type(b))
    print(b)
    print(type(args))
    print(args)
    sum = 0
    for i in args:
        sum += i
    
    print(sum)
    print(type(kwargs)) 
    print(kwargs)
    
add(2,3,4,5,5,1,1,1,1,1,1,1,1,5,5,7,7,7,7,7,7,7,7,999,7, val1 =200)
```

```python
def add(a, b,c=3, *args, **kwargs):
    print(type(a))
    print(a)
    print("-"*10)
    print(type(b))
    print(b)
    print(type(args))
    print(args)
    sum = 0
    for i in args:
        sum += i
    
    print(sum)
    print(type(kwargs)) 
    print(kwargs)
    
add(2,3,4,5,5,1,1,1,1,1,1,1,1,5,5,7,7,7,7,7,7,7,7,999,7, val1 =200, val2=300)
```

# *args is a TUPLE
# **kwargs is a DICTIONARY

```python
def add(a, b,c=3, *args, **kwargs):
    print(type(a))
    print(a)
    print("-"*10)
    print(type(b))
    print(b)
    print(type(args))
    print(args)
    sum = 0
    for i in args:
        sum += i
    
    print(sum)
    print(type(kwargs)) 
    print(kwargs)
    
a = {'val1': 200, 'val2': 300}

add(2,3,4,5,5,1,1,1,1,1,1,1,1,5,5,7,7,7,7,7,7,7,7,999,7, **a)
```

# Example of *args and **kwargs

```python
def order_pizza(a, *args, **kwargs):
    print(a)
    print(args)
    print(kwargs)

order_pizza(20, "Margherita", "Extra Cheese", "Mushrooms", size="Large", crust="Thin")
```
