```python
try:
    with open(r"C:\Users\LazarThomas\Desktop\sample.txt", "r+") as f:
        a = f.tell1();
        print(a);
        f.write("lazar");
        print(f.tell);

except FileNotFoundError:
    print("file not found");
except AttributeError:
    print("AttributeError");
```
